```python
import sqlite3
import re
import gensim
import logging
import nltk.data
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from gensim.models import word2vec

import warnings
warnings.filterwarnings('ignore')
```

```python
conn = sqlite3.connect('dom.db')
cur = conn.cursor()
cur.execute("""SELECT normal_form, POS FROM words""")
words = cur.fetchall()
conn.close()
```

```python
words[0:10]
```

```python
words_to_file = []
for one_word in words:
    if one_word[1] != None:
        words_to_file.append('_'.join(one_word)) 
```

```python
with open('dom_w2v_lemm.txt', 'w', encoding='utf-8') as f:
    f.write(' '.join(words_to_file))
```

Что делать с именами персонажей?

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
with open('dom.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
words = nltk.word_tokenize(text)
print(words[0:30])
```

```python
names = []
punct = '.!?'
for i, word in enumerate(words):
    if word.istitle() == True and words[i-1] not in punct and word not in names:
        names.append(word)
```
